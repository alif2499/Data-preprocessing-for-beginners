```python
#iporting neccesary libraries
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
#load dataset
dataset = pd.read_csv('../input/csvdata/Data.csv')
```

```python
#show first 5 data from dataset
dataset.head(5)
```

```python
#extract the independent variables
X = dataset.iloc[:, :-1].values
#extract the dependent variables
Y = dataset.iloc[:, -1].values
```

```python
from sklearn.impute import SimpleImputer                               #for handling the missing data
imputer = SimpleImputer(missing_values = np.nan, strategy = 'mean')    #declaring the imputer object
imputer = imputer.fit(X[:,1:3])                                        #fitting the train data
X[:,1:3] = imputer.transform(X[:,1:3])                                 #transforming the missing data value to mean data value
```

```python
#Encoding the categorical variables
from sklearn.preprocessing import LabelEncoder, OneHotEncoder
#labelencoder = LabelEncoder()
#X[:,0] = labelencoder.fit_transform(X[:,0])
#onehotencoder = OneHotEncoder(categorical_features = [0])
#X = onehotencoder.fit_transform(X).toarray()
from sklearn.compose import ColumnTransformer
ct = ColumnTransformer([("Country", OneHotEncoder(), [0])], remainder = 'passthrough')
X = ct.fit_transform(X)
```

```python
#encoding the dependent variable. No need of oneHotEncoding as it is dependent variable
labelencoder_y = LabelEncoder()
Y = labelencoder_y.fit_transform(Y)
```

```python
#splitting the dataset into training and test set
from sklearn.model_selection import train_test_split
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size = 0.2, random_state = 0)
```

```python
#Feature Scaling
from sklearn.preprocessing import StandardScaler
sc_X = StandardScaler()
X_train = sc_X.fit_transform(X_train)
X_test = sc_X.transform(X_test)
```
